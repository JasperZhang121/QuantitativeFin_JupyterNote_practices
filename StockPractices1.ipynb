```python
df = pd.read_csv('./maotai.csv')
```

```python
df.drop(labels='Unnamed: 0',axis=1, inplace= True)
```

```python
# convert date to timeseries
df['date'] = pd.to_datetime(df['date'])
```

```python
df.info()
```

```python
df.set_index('date',inplace=True)
```

```python
df.head()
```

```python
# get (closing - opening)/opening > 0.03
df.loc[(df['close']-df['open'])/df['open']>0.03].index
```

```python
# get (T.opening- T-1. closing -)/T-1.closing <- 0.02

df['close'].shift(1)
```

```python
(df['open']-df['close'].shift(1))/df['close'].shift(1)<-0.02
```

```python
df.loc[(df['open']-df['close'].shift(1))/df['close'].shift(1)<-0.02].index
```

```python
new_df = df['2010-01':'2020-02']
new_df
```

```python
# buy stocks
df_monthly = new_df.resample('MS').first()
df_monthly
```

```python
# cost 
cost = df_monthly['open'].sum()*100
cost
```

```python
new_df.resample('A').last()
```

```python
df_yearly = new_df.resample('A').last()[:-1]
df_yearly
```

```python
gain = df_yearly['open'].sum()*1200
gain
```

```python
net = gain - cost + 200*new_df['close'][-1]
net
```
