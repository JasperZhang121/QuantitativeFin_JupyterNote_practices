## Concatenate Operations

- Adding numpy array vertially and horizontally

```python
# concatenate numpy array horizontally, axis=1
np.concatenate((arr,arr),axis=1)
```

```python
# concatenate numpy array vertially, axis = 0
np.concatenate((arr,arr), axis=0)
```

What if not exact shape?

```python
test1 = np.array([[1,2,3],[1,2,3]])
```

```python
test2 = np.array([[1,1,1]])
```

```python
np.concatenate((test1,test2), axis=0)
```

Conclusion: must be in same dimensions for two arrays, but not necessarily to have identical shape.
